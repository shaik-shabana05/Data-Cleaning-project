```python
budget_revenue_data = data.loc[(data.budget != 0) & (data.revenue != 0),:]
budget_revenue_data.shape
```

```python
budget_revenue_data['profit']=data.revenue-data.budget
```

```python
budget_revenue_data['roi']=(budget_revenue_data.revenue-budget_revenue_data.budget)/budget_revenue_data.budget
```

```python
budget_revenue_data.loc[budget_revenue_data['roi']<-1,'roi']=-1
budget_revenue_data.loc[budget_revenue_data['roi']>10,'roi']=10
```

```python
#generate column new year
budget_revenue_data['release_date']=pd.to_datetime(budget_revenue_data['release_date'],errors='coerce#')
budget_revenue_data['release_year']=budget_revenue_data['release_date'].dt.year
```

```python
#genres count per movie
genre_cols=['Crime','Adventure','Family',
       'Foreign', 'War', 'TV Movie', 'Horror', 'Romance', 'Animation', 'Drama',
       'Comedy', 'Thriller', 'Action', 'History', 'Mystery', 'Fantasy',
       'Documentary', 'Science Fiction', 'Western', 'Music']
budget_revenue_data['genre_count']=budget_revenue_data[genre_cols].sum(axis=1).astype('int64')
```

```python
budget_revenue_data.shape
```

```python
#to check if a movie is blockbuster or not (10cr)
budget_revenue_data['blockbuster']=(budget_revenue_data.revenue>100_000_000).astype('int64')
#budget_revenue_data['blockbuster']=budget_revenue_data['blockbuster'].astype(int)
```
